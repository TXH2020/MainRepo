##Q1. One ethics study suggested that 84% of U.S. companies have an ethics code.From a random sample of 15 companies, what is the probability that at least 10 have an ethics code?
###Here, there are two possibilities for each trial: having an ethics code or not. Each event is independent of each other. Thus it is an example of binomial distribution.
###Here we have to find the probability of k>=10. Thus the answer is:
###1-P(k<10)
###where P(k<10) = P(k<=10) - P(k=10)

```python
print("Answer to Q1:")
result=1-(st.binom.cdf(k=10,n=15,p=0.84)-st.binom.pmf(k=10,n=15,p=0.84))
print(result)
```

##Q3. Suppose the amount of time it takes to assemble a plastic module ranges from 27 to39 seconds and that assembly times are uniformly distributed. Describe the distribution.What is the probability that a given assembly will take between 30 and 35 seconds?
###Here, the assembly times are uniformly distributed i.e., the probability density function will remain constant over the specified interval. Thus it is an example of **uniform distribution**
i2.JPG
###In this case, x2=35,x1=30,b=39,a=27
###The probability can also be shown as
###P(x<=35)-P(x<=30)


```python
import numpy as np
result=st.uniform.cdf(35,loc=27,scale=12)-st.uniform.cdf(30,loc=27,scale=12)
print("Answer for Q3:",result)
```
