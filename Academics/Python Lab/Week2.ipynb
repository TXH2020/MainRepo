```python
import ast
def Sum(l):
  return sum(l)
def Multiply(l):
  if (len(l)==0):
    return 0
  if(len(l)==1):
    return 1
  else:
    return l[-1]*Multiply(l[:len(l)-1])
x=ast.literal_eval(input("Enter a list"))
print(Sum(x))
print(Multiply(x))
```

```python

def displaySecondMax(l):
  l1=l.copy()
  l1.remove(max(l1))
  return max(l1)
print(displaySecondMax([3,3]))
```

```python
def rec_lin_search(l,k,b):
  if(len(l)==0 or len(l)<=b):
    return -1
  elif(k==l[b]):
    return b+1
  else:
    return rec_lin_search(l,k,b+1)
rec_lin_search([],5,0)
```

```python
def binsearch(l,k):
  if(len(l)==0):
    print("list is empty")
    return
  l=sorted(l)
  print("Sorted array is:",l)
  b=0;e=len(l)-1
  while(b<=e):
    m=(b+e)//2
    if(l[m]==k):
      print("Element ",k," found at position:",m+1)
      return
    elif(k>l[m]):
      b=m+1
    else:
      e=m-1
  print("Element ",k," not found")
binsearch([5,2,3,1,4],6)
```

```python
def displayFibonacci():
  print("The fibonacci nos. between 0 and 100 are:\n")
  a=0;b=1
  print(a,b,end=',',sep=',')
  while(True):
    c=a+b
    if(c>100):
      break
    print(c,end=',')
    a=b
    b=c
displayFibonacci()
```

```python
import ast
def findQuadraticRoots():
  a,b,c=ast.literal_eval(input("Enter the coefficients of the quadratic equation(a,b,c)"))
  if(a==0):
    print("Not a quadratic equation")
    return
  d2=b*b-4*a*c
  if(d2>=0):
    r1=(-b+d2**.5)/(2*a)
    r2=(-b-d2**.5)/(2*a)
    print("Root1=%.2f"%r1)
    print("Root2=%.2f"%r2)
  else:
    d2=-d2
    r1=-b/(2*a)
    r2=(d2**.5)/(2*a)
    print("Root1=%.2f+%.2fi"%(r1,r2))
    print("Root1=%.2f-%.2fi"%(r1,r2))
findQuadraticRoots()
```
