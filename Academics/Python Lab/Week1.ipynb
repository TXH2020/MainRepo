```python
import ast
t=ast.literal_eval(input("Enter 4 subject marks:(a,b,c....)"))
m=int(sum(t)/4)
if(m>=91 and m<=100):
  print("Grade S")
elif(m>=81 and m<=90):
  print("Grade A")
elif(m>=71 and m<=80):
  print("Grade B")
elif(m>=61 and m<=70):
  print("Grade C")
elif(m>=51 and m<=60):
  print("Grade D")
elif(m>=41 and m<=50):
  print("Grade E")
else:
  print("Grade F")
```

```python
import ast
a,b=ast.literal_eval(input("Enter an interval(a,b):"))
for i in range(a,b+1):
  k=1
  if(i==1):
    k=0
  for j in range(2,i//2+1):
    if(i%j==0 and i!=j):
      k=0
      break
  if(k==1):
    print(i,end=',')
```
